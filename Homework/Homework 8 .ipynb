```python
def assignTeams(inputfile,n):
    import pandas as pd
    input1 = pd.read_excel(inputfile,sheet_name='Students',index_col=0)
    input2 = pd.read_excel(inputfile,sheet_name='Parameters',index_col=0)
    A = input1
    W = input2.loc['Weights']
    L = input2.loc['L']
    U = input2.loc['U']
    J = [i for i in range(1,n+1)]
    I = input1.index
    K = input2.columns
    mod = Model()
    x = mod.addVars(I,J,vtype=GRB.BINARY)
    s = mod.addVars(K,lb=0)
    t = mod.addVars(K,lb=0)
    mod.setObjective(sum(W[k]*(s[k]+t[k]) for k in K))
    for i in I:
        mod.addConstr(sum(x[i,j] for j in J)==1)
    for j in J:
        for k in K:
            mod.addConstr(sum(x[i,j]*A.loc[i,k] for i in I) >= L.loc[k]-s[k])
            mod.addConstr(sum(x[i,j]*A.loc[i,k] for i in I) <= U.loc[k]+t[k])
    mod.setParam('outputflag', False)
    mod.optimize()
    
    df=pd.DataFrame(columns=['Team'],index=I)
    for i in I:
        for j in J:
            if x[i,j].x:
                df.loc[i,'Team'] = j
    return df,mod.objval
```

```python
df,objval=assignTeams('HWK8-Q5-input-2.xlsx',10)
print('Optimal objective value',objval)
df.sort_values(by='Team')
```
